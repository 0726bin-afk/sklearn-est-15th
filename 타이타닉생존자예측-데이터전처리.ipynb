# 4. 결측치 처리 (수치형, 문자형)

**중앙값(median)**<br>
데이터: [3,1,2,4,5,100] -> 정렬 [1,2,3,4,5,100] -> 중앙값 4

**평균(mean)**<br>
데이터: [3,1,2,4,5,100] -> 평균 21.66666666

```python
# fillna()를 사용하여 결측치 채우기 (평균값)
train_df[['Age']].fillna(train_df[['Age']].median(), inplace = False)
```

```python
# sklearn의 simple imputer를 사용한 결측치 처리
from sklearn.impute import SimpleImputer

# 1. simple imputer 생성 
# strategy는 mean, median, most_frequent, constant
# 수치형 데이터는 4자기 모두 사용가능
# 문자열 데이터는 most_frequent, constant 사용
imputer = SimpleImputer(strategy = 'mean')

# 2. fit_transform으로 결측치 처리
train_df[['Age']] = imputer.fit_transform(train_df[['Age']])
```

```python
train_df.isnull().sum()
```

```python
# 최빈값이 무엇인지 확인
train_df['Embarked'].value_counts()
```

```python
# Embarked 결측치 처리 (imputer로 최빈값(most_frequent)으로 채우기)

# 1. imputer 생성
imputer = SimpleImputer(strategy = 'most_frequent')

# 2. fit_transform으로 결측치 처리
train_df[['Embarked']] = imputer.fit_transform(train_df[['Embarked']])
```

```python
train_df.isnull().sum()
```

```python
# 최빈값으로 결측치 지워졌는지 확인
train_df['Embarked'].value_counts()
```

# 5. 레이블 인코딩
* 인공지능 모델은 문자열을 처리할 수 없습니다. 따라서 문자열 데이터를 숫자로 변환하는 것이 필요합니다. 
* 이 과정을 레이블 인코딩이라고 합니다.



```python
# train_df에서 레이블 인코딩을 해야하는 대상은? ['Sex']. ['Embarked']
# sklearn에는 LabelEncoder가 있다. 'male', 'female'을 0, 1로 변환

from sklearn.preprocessing import LabelEncoder

# 1. 레이블 인코더를 생성
label_encoder = LabelEncoder()

# 2. fit_transform으로 레이블 인코딩을 적용
# fit은 데이터를 확인 'male', 'female'을 0, 1로 변환
train_df['Sex'] = label_encoder.fit_transform(train_df['Sex'])
# fit은 데이터를 확인 'S', 'C', 'Q'를 0, 1, 2로 변환
train_df['Embarked'] = label_encoder.fit_transform(train_df['Embarked'])
```

```python
train_df.head()
```

# 7. 정규화(Normalization), 표준화(Standardization)

```python
movie = {'naver': [2, 4, 6, 8, 10], 
         'netflix': [1, 2, 3, 4, 5]
         }
```

```python
movie = pd.DataFrame(data=movie)
movie
```

### MinMaxScaler

```python
# sklearn에는 정규화를 위한 함수로 MinMaxScaler가 있다.
from sklearn.preprocessing import MinMaxScaler
# min = 0, max = 5

# 1. MinMaxScaler를 생성 (min=0, max=1)
min_max_scaler = MinMaxScaler(feature_range=(0,1))

# 2. fit_transform으로 데이터를 변환
movie2 = min_max_scaler.fit_transform(movie)
movie2
```

- 여러개의 컬럼이 존재할 때 서로 다른 데이터의 특성(단위가)이 다른 데이터를 비교

    - 예시를 집값 데이터로 들면,<br>
    집의 크기(m^2), 방의 개수(개), 지하철 역과의 거리(km), 마트와의 거리(km)<br>
    집의 크기: 50 ~ 130<br>
    방의 개수: 1 ~ 5<br>
    지하철 역과의 거리: 0.1 ~ 10<br>
    마트와의 거리: 0.1 ~ 9<br>


```python
import pandas as pd

# 1. 데이터 생성 (딕셔너리 형태)
data = {
    '집의 크기(m2)': [85.5, 120.2, 55.0, 105.8, 72.4],
    '방의 갯수(개)': [3, 4, 1, 3, 2],
    '지하철역과의 거리(km)': [0.5, 2.1, 0.2, 5.4, 1.1],
    '마트와의 거리(km)': [1.2, 0.8, 4.5, 2.3, 0.3]
}

# 2. 데이터프레임 생성
df = pd.DataFrame(data)

# 3. 결과 출력
print(df)
```

```python
# df를 정규화(MinMaxScaler)
# 1. MinMaxScaler를 생성 (min=0, max=1)
min_max_scaler = MinMaxScaler(feature_range=(0,1))

# 2. fit_transform으로 데이터를 변환
df2 = min_max_scaler.fit_transform(df)
```

```python
df2
```

## 표준화 (Standard Scaling)

* 평균이 0이고, 표준 편차가 1이 되도록 변환


```python
# 표준화를 위한 샘플 데이터셋 생성
x = np.arange(10)

# 이상치(outlier)를 추가
x[9] = 1000
x
```

```python
x.mean(), x.std()
```

```python
# 1차원 -> 2차원
x = x.reshape(-1,1)
x
```

```python
# sklearn에서 StandardScaler
from sklearn.preprocessing import StandardScaler

# 1. StandardScaler 생성
standard_scaler = StandardScaler()

# 2. fit_transform
standard_scaled = standard_scaler.fit_transform(x)
standard_scaled
```

```python
# 실제로 평균이 0이고, 표준 편차가 1이 되었는가? 확인
# 4x10^-17 -> 거의 0, 표준 편차 1
standard_scaled.mean(), standard_scaled.std()
```
